```python
#for more word clouds here it is the resource https://github.com/amueller/word_cloud

import numpy as np
from PIL import Image
from os import path
import matplotlib.pyplot as plt
import random

from wordcloud import WordCloud, STOPWORDS


def grey_color_func(word, font_size, position, orientation, random_state=None,
                    **kwargs):
    return "hsl(0, 0%%, %d%%)" % random.randint(60, 100)

d = "/Users/aybuke/Documents"

# read the mask image taken from 
#https://www.123rf.com/photo_20161949_3d-white-retro-astronaut-with-spaceship-and-thumb-up-isolated-white-background.html


mask = np.array(Image.open(path.join(d, "dragon4.png")))


# 400+ unique twitter users based on the way they express themselves on their profile

text = open( '/Users/aybuke/Documents/U_desc.txt', encoding='latin-1').read()


# add specific stopwords
stopwords = set(STOPWORDS)
stopwords.add("html")

wc = WordCloud(max_words=1000, mask=mask, stopwords=stopwords, margin=10,
               random_state=1).generate(text)
# store default colored image
default_colors = wc.to_array()
plt.title("Who are the participants?")
plt.imshow(wc.recolor(color_func=grey_color_func, random_state=3),
           interpolation="bilinear")
wc.to_file("outPut_usersProfiles.png")
plt.axis("off")
plt.figure()
plt.show()
```
